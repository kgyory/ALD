```python
import matplotlib.pyplot as plt
import math

X = range(1, 50)

Y1 = [math.log2(value) for value in X]
Y2 = [value for value in X]
Y3 = [math.log2(value) * value for value in X]
Y4 = [value ** 2 for value in X]
Y5 = [value ** 3 for value in X]

plt.plot(X,Y1, label = "O(log n)")
plt.plot(X,Y2, label = "O(n)")
plt.plot(X,Y3, label = "O(n log n)")
plt.plot(X,Y4, label = "O(n^2)")
plt.plot(X,Y5, label = "O(n^3)")

plt.ylim(0, 500)
plt.xlim(1, 30)

plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import math

X = range(1, 50)

Y1 = [math.log2(value) for value in X]
Y2 = [value for value in X]
Y3 = [math.log2(value) * value for value in X]
Y4 = [value ** 2 for value in X]
Y5 = [value ** 3 for value in X]

plt.figure(figsize = [15, 10])
plt.suptitle('important functions used in the analysis of algorithms', fontsize=16)

grid = plt.GridSpec(3, 6, wspace = 1, hspace = 0.5)

plt.subplot(grid[0, :3], title = "O(log n)")
plt.plot(X,Y1)

plt.subplot(grid[0, 3:], title = "O(n)")
plt.plot(X,Y2, color = "red")

plt.subplot(grid[1:, :2], title = "O(n log n)")
plt.plot(X,Y3, color = "green")

plt.subplot(grid[1:, 2:4], title = "O(n^2)")
plt.plot(X,Y4, color = "limegreen")

plt.subplot(grid[1:, 4:], title = "O(n^3)")
plt.plot(X,Y5, color = "brown")

plt.show()
```
